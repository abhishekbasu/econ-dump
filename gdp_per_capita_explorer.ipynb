```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Data source
Download the data from the World Bank source [here](https://api.worldbank.org/v2/en/indicator/NY.GDP.PCAP.CD?downloadformat=csv).

```python
data = pd.read_csv("~/Downloads/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_5994720/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_5994720.csv", skiprows=4)
```

```python
YEAR_MIN = 1960
YEAR_MAX = 2022
```

```python
def plot_gdp(data: pd.DataFrame, country_names: list[str]):
    assert len(country_names) > 0, "You need atleast one country to plot."
    assert len(country_names) < 7, "Too many countries to plot."
    colors = ["darkred", "darkblue", "goldenrod", "darkgreen", "orange", "purple"]
    years = list(map(str, range(YEAR_MIN, YEAR_MAX+1)))
    for i, country_name in enumerate(country_names):
        gdp_values = data.loc[data["Country Name"] == country_name, years].values.reshape(63,)
        plt.plot(years, gdp_values, color=colors[i], label=country_name)
        plt.xticks(years[::5])
    plt.title("GDP per capita for countries (1960 - 2023)")
    plt.xlabel("Year")
    plt.ylabel("GDP per capita ($)")
    plt.legend()
    plt.show()
```

```python
plot_gdp(data, ["China", "Mexico"])
```
